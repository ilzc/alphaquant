```python
from empyrical import cum_returns, annual_return, sharpe_ratio, max_drawdown

import pandas as pd
import empyrical as em
import matplotlib.pyplot as plt
df=pd.read_csv("results.csv")
df=df[['time','pnl','returns']]

# df["time"]=pd.to_datetime(df["time"])
# df=df.set_index(df["time"])
df.plot(x="time",y="pnl")
# plt.show()
df["time"]=pd.to_datetime(df["time"])
df=df.set_index(df["time"])
returns=df['returns']
ann_return = annual_return(returns)   # 计算累计收益率
cum_return_list = cum_returns(returns)
   # 计算sharp ratio
sharp = sharpe_ratio(returns)
  # 最大回撤
max_drawdown_ratio = max_drawdown(returns)
print("年化收益率 = {:.2%}, 累计收益率 = {:.2%}, 最大回撤 = {:.2%}, 夏普比率 = {:.2f} ".format
        (ann_return, cum_return_list[-1], max_drawdown_ratio, sharp))


```
